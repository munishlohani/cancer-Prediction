```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('breast_cancer.csv')
df
```

```python
from sklearn.model_selection import train_test_split
df_train,df_test=train_test_split(df,train_size=0.7,test_size=0.3,random_state=1)
```

```python
y_train=df_train.pop('Class')
x_train=df_train




```

```python
from sklearn.linear_model import LinearRegression
ln=LinearRegression()
model=ln.fit(x_train,y_train)



```

```python
# Testing
data=df_test.iloc[16:17]
actual_class=data.pop('Class')
prediction=model.predict(data)
print('Actual class= ',actual_class)
print('predicted Class= ',prediction[0])
```

```python
df
```

```python
y_test=df_test.pop('Class')
x_test=df_test
```

```python
prediction=model.predict(x_test)
```

```python
df_predict=pd.DataFrame({'Actual Class':y_test,'Predicted Class':prediction})
```

```python
df_predict
```

## Calculating Error

```python
from sklearn.metrics import mean_squared_error
```

```python
error=mean_squared_error(df_predict['Actual Class'],df_predict['Predicted Class'])
error
```

## Predicting Cancer Class and forming a Data Frame

```python
x=int(input('Enter Data Frame elements= '))



dataframe1=pd.DataFrame({'Clump Thickness':[],'Uniformity of Cell Size':[],'Marginal Adhesion':[],'Single Epithelial Cell Size':[],
                           'Single Epithelial Cell Size':[],
                            'Bare Size':[],
                            'Bland Chromation':[],
                            'Normal Nucloli':[],
                            'Mitosis':[],
                            'Class':[]
                            
                           
                           })
for i in range(x):
    a=int(input('Enter Clump Thickness= '))
    b=int(input('Enter Uniformity of Cell Size= '))
    c=int(input('Enter Uniformity of Cell Shape= '))
    d=int(input('Enter Marginal Adhesion= '))
    e=int(input('Enter Single Epithelial Cell Size= '))
    f=int(input('Enter Bare Size= '))
    g=int(input('Enter Bland Chromation= '))
    h=int(input('Enter Normal Nucloli= '))
    i=int(input('Enter Mitosis= '))
    x_val=np.array([a,b,c,d,e,f,g,h,i])
    cl=model.predict(x_val.reshape(1,-1))
    dataframe=pd.DataFrame({'Clump Thickness':[a],'Uniformity of Cell Size':[b],'Marginal Adhesion':[c],'Single Epithelial Cell Size':[d],
                           'Single Epithelial Cell Size':[e],
                            'Bare Size':[f],
                            'Bland Chromation':[g],
                            'Normal Nucloli':[h],
                            'Mitosis':[i],
                            'Class':cl[0]
                            
                           
                           })
    dataframe1=dataframe1.append(dataframe)
dataframe1
```
